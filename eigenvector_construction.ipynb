Say our eigenvectors are specified in $basis$ , and we have weights $a_n$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a  = np.array([ 0.38626417, -0.68651368, -0.44271343, -0.42837341]) #Normalized 
```

```python
def eigenvector(L, a, x):
    N = len(a)
    y = np.zeros(len(x))
    for n in range(N):
        y += a[n] * np.sin(n*np.pi*x/L)
        
    return y
```

```python
x = np.linspace(-1,1,100)
```

```python
y  = eigenvector(1 , a, x)
```

```python
plt.plot(x,y)
```
